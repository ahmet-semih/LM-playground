```python
path="/Users/ahmetsemihgumus/LM-playground/ahmet/llama"
#merge all files ending with txt
import os
with open("merged_pretrain_data.txt", "w", encoding="utf-8") as outfile:
    for filename in os.listdir(path):
        if filename.endswith(".txt"):
            with open(os.path.join(path, filename), "r", encoding="utf-8") as infile:
                outfile.write(infile.read() + "\n")
```

```python
#remove extra newliine space from merged_pretrain_data.txt
with open("merged_pretrain_data.txt", "r", encoding="utf-8") as f:
    lines = f.readlines()
with open("merged_pretrain_data_cleaned.txt", "w", encoding="utf-8") as f:
    for line in lines:
        if line.strip():  # only write non-empty lines
            f.write(line)
```
